```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, GridSearchCV
```

https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic)


```python
#Importation des données
df=load_breast_cancer()
df
```

```python
print(df)
```

```python
#Variable cible
y=df.target
print(y)
```

```python
X=df.data
print(X)
```

```python
X.shape
```

```python
#Initiallisation de KNN avec ses hyperparamètre par défaut
knn=KNeighborsClassifier()
```

```python
#5-fold cross-validation
```

```python
cv_scores=cross_val_score(knn,X,y,cv=5,scoring='precision')
```

```python
#Affichage des scores
print(cv_scores)
```

```python
#Score moyen arrondi à deux chiffres après la virgule
round(np.mean(cv_scores),2)
```

Grid Search

```python
#Les hyperparamètres
```

```python
#Affichage de knn
print(knn)
```

https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html

```python
#Dictionnaire des hyperparamètres avec leurs valeurs
grid={'n_neighbors':[1,3,5,7,9,10],
      'weights':['uniform','distance'],
      'algorithm':['auto','ball_tree','kd_tree','brute'],
      'metric':['euclidiean','manhattan'],
      'leaf_size':[10,20,30,40,50]
    
}
```

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html

```python
#Estimateur
model=GridSearchCV(estimator=KNeighborsClassifier(),param_grid=grid,scoring='precision',cv=5)
```

Le nombre de combinaison (nombre de modèle) avec nos hyperparamètre dans grid est 6*2*4*2*5=480. Et cv=5 modèles sont créé pour chaque combinaison. Alors le nombre total de modèle créé est :780*5=2400 modèles

```python
#Recherche du meilleur modèle
model.fit(X,y)
```

```python
model.cv_results_
```

```python
#convertion du dictionnaire en une dataframe
results=pd.DataFrame(model.cv_results_)
```

```python
#On s'interresse à 'params' et 'mean_test_score'
best=results[['params','mean_test_score']].sort_values(by='mean_test_score',ascending=False).head()
best
```

```python
#On remarque que les 5 meilleurs modèles ont le meme score avec des hyperparamètres différents
```
